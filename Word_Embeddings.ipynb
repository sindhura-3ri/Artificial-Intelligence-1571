```python
from warnings import filterwarnings
filterwarnings('ignore')

import gensim

from gensim.models import Word2Vec # word to vector conversion
from nltk.tokenize import word_tokenize,sent_tokenize
from nltk.corpus import gutenberg
```

```python
gutenberg.fileids()
```

```python
text = gutenberg.raw('shakespeare-hamlet.txt')
text
```

```python
# will create sentence tokens
sentences = sent_tokenize(text)

# will create word tokens
word_tokens = [word_tokenize(sent.lower()) for sent in sentences]
word_tokens
```

```python
len(word_tokens)
```

# Embeddings: Word will convert into vector format on basis of multiple dimensions.

    Dimensions: For human understanding, consider them as properties/features/characteristic of word.

    For example:
    Tiger : vector representation - **100** embeddings(properties/characterstics of Tiger)
    1) Wild animal 2) Feline 3) Canine 4) Domestic animal 5)birds 6)living being 7) non-living being
    vector representation : [0.9,0.87,0.45,.......] (**100**)

## Continuous Bag of Words(CBOW) and Skip Gram Model(SGM)

    CBOW => multiple inputs, it will predict one word as output

    SGM => single input, it will predict mutliple words as output

    We will be using the CBOW architecture:

    min_count=1: Includes words that appear at least once.
    vector_size=100: Generates word vectors of 100 dimensions.
    window=5: Considers a context window of 5 words before and after the target word.
    sg=0: Uses CBOW model (default setting).

    For Skip Gram:
    In above input, change the sg=1.


    Example:
    Context Window

    consider 5 words before and 5 words after the target word

    The movie is amazing. the acting done in the movie was top notch.

```python
# CBOW
model1 = gensim.models.Word2Vec(word_tokens,min_count=1,vector_size=100,window=5)

# Skip Gram
model2 = gensim.models.Word2Vec(word_tokens,min_count=1,vector_size=100,window=5,sg=1)
```

```python
print("CBOW")
print("Cosine similarity between william and shakespeare",model1.wv.similarity('william','shakespeare'))
print("=========")
print("Skip Gram")
print("Cosine similarity between william and shakespeare",model2.wv.similarity('william','shakespeare'))
```

```python
print("CBOW")
print("Cosine similarity between Barnardo and Francisco",model1.wv.similarity('barnardo','francisco'))
print("=========")
print("Skip Gram")
print("Cosine similarity between Barnardo and Francisco",model2.wv.similarity('barnardo','francisco'))
```

```python
import gensim.downloader as api

# Load pre-trained GloVe model (50-dimensional vectors, trained on Wikipedia + Gigaword)
glove = api.load("glove-wiki-gigaword-50") # When loaded, glove behaves like a dictionary: keys = words, values = vectors.

# Similarity between related words-computes cosine similarity between the two word vectors.
print("Similarity between king and queen:", glove.similarity("king", "queen"))
print("Similarity between paris and france:", glove.similarity("paris", "france"))

# Similarity between unrelated words-computes cosine similarity between the two word vectors.
print("Similarity between king and banana:", glove.similarity("king", "banana"))
print("Similarity between computer and love:", glove.similarity("computer", "love"))

# Famous analogy: king - man + woman ≈ queen
result = glove.most_similar(positive=["king", "woman"], negative=["man"], topn=1)
print("Analogy king - man + woman ->", result)
```
