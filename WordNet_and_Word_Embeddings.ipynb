## WordNet :WordNet is a lexical(word vocabulary) database of English that groups words into synsets (sets of synonyms) and defines relationships between them such as:
	• Synonym (similar meaning)
	• Antonym (opposites)
	• Hypernym/Hyponym (general–specific hierarchy)
	• Meronym/Holonym (part–whole relationship)
It becomes necessary (or useful) in tasks where you need semantic meaning of words, not just raw text.

## Text Similarity: Measures of Similarity
1) Wup similarity
2) Path Similarity

```python
s1 = wn.synsets('awesome')[0]
s2 = wn.synsets('good')[0]
```

```python
s1
```

```python
s2
```

```python
s1.definition()
```

```python
s2.definition()
```

```python
s1.path_similarity(s2) # shortest distance
```

```python
s1.wup_similarity(s2) # how close are they to the ancestoral category. do they share same or not?
```

```python
s3=wn.synsets("dog")[0]
s4= wn.synsets("cat")[0]
print(s3.definition(),s4.definition(),sep='\n')
```

```python
s3.path_similarity(s4)
```

```python
s3.wup_similarity(s4)
```

## Word Sense Disambiguition

```python
pip install pywsd
```

```python
from pywsd.lesk import simple_lesk

sentence1 = "He deposited the cheque at the bank"
sentence2 = "He sat on the bank of the river"

sl_wsd1 = simple_lesk(sentence1, 'bank')
sl_wsd2 = simple_lesk(sentence2, 'bank')

print(sentence1)
print(f"Disambigued Sense: {sl_wsd1.definition()}")

print("===================",sentence2,sep='\n')
print(f"Disambigued Sense: {sl_wsd2.definition()}")
```
