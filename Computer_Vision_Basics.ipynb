```python
import cv2
```

```python
path1 = r"/content/3RI logo.png"
img = cv2.imread(path1)
img
```

```python
import matplotlib.pyplot as plt
plt.imshow(img)
```

## Files are loaded in BGR format instead of RGB format

RGB [red,green,blue] = [90,7,100]
BGR = [90,7,100]

```python
img2 = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img2
```

```python
plt.imshow(img2)
```

```python
img2.shape
```

## Load a colored image in black and white, load it in grayscale format

```python
img3 = cv2.imread(path1,cv2.IMREAD_GRAYSCALE)
img3 # (78,88,1)
```

```python
img3.shape
```

```python
plt.imshow(img3)
```

```python
plt.imshow(img3,cmap='gray')
```

```python
import seaborn as sns
plt.figure(figsize=(25,25))
sns.heatmap(img3,annot=True,fmt='d')
plt.show()
```
