## Data Preprocessing
Feature scaling: we will bring all the numbers to a standard scale by dividing them with 255.

```python
xtrain[0]
```

```python
xtrain_scl = xtrain.astype("float32")/255
xtest_scl = xtest.astype("float32")/255
```

```python
from keras.utils import to_categorical
ytrain_scl = to_categorical(ytrain)
ytest_scl = to_categorical(ytest)
```

## expand the dimensions to tell keras about the color of the image

1) RGB - we dont expand dimensions for RGB as it is mentioned in the shape of colored images.
2) Grayscale - needs to be done for grayscale as it is NOT mentioned in the shape of grayscale images

```python
xtrain_scl.shape
```

```python
import numpy as np
xtrain_scl = np.expand_dims(xtrain_scl,axis=-1)
xtest_scl = np.expand_dims(xtest_scl,axis=-1)
```
