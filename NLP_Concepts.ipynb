## TOkenization

```python
data = '''
Data analysis is the process of inspecting, cleansing, transforming, and modeling data with the goal of discovering useful information, informing conclusions, and supporting decision-making.
[1] Data analysis has multiple facets and approaches, encompassing diverse techniques under a variety of names, and is used in different business, science, and social science domains.
[2] In today's business world, data analysis plays an important role in making decisions more scientific and helping businesses operate more effectively.
[3] It is widely used in fields such as business analytics, healthcare, and artificial intelligence to extract meaningful insights from data.
'''
print(data)
```

## Text cleaning

```python
# 1) convert all the data into lowercase
data = data.lower()

# 2) remove punctuation marks. remove numbers and also the special characters
import re
pattern = r'[^a-z. ]'
data = re.sub(pattern,'',data)
print(data)
```

## Word tokenization

1) Unigrams
2) Bigrams : 2 words in a token
3) Trigrams : 3 words in a token
4) N grams : n words in a token

Unigrams : ['data', 'analysis', 'is', 'the', 'process', 'of',]

Bigrams : ['data analysis','analysis is','is the','the process','process of',..]

Trigrams : ['data analysis is','analysis is the','is the process','the process
of','process of ...]

N grams -5: ['data analysis is the process','analysis is the process of',...]


```python
bi_tokens = list(nltk.bigrams(word_tokens))
print(bi_tokens)
```

```python
tri_tokens = list(nltk.trigrams(word_tokens))
print(tri_tokens)
```

```python
ntokens = list(nltk.ngrams(word_tokens,5))
print(ntokens)
```

## Stemming and Lemmatization

Stemming

image.png

**Porter stemmer**

```python
from nltk.stem import PorterStemmer

words = ["easily","walking","happiness","given","marked","inspecting"]

ps = PorterStemmer()

ps_words = [ps.stem(word) for word in words]
print(ps_words)
```

**Snowball Stemmer** : supoorts stemming technique for other languages also

```python
from nltk.stem import SnowballStemmer

words = ["easily","walking","happiness","given","marked","inspecting"]

ss = SnowballStemmer("english")

ss_words = [ss.stem(word) for word in words]
print(ss_words)
```

**Lancaster Stemmer**

```python
from nltk.stem import LancasterStemmer

words = ["easily","walking","happiness","given","marked","inspecting"]

ls = LancasterStemmer()

ls_words = [ls.stem(word) for word in words]
print(ls_words)
```

## Lemmatization : brings the word to root form on the basis of part of speech tags

image.png

image.png



```python
from nltk.stem import WordNetLemmatizer

word_lem = WordNetLemmatizer()

words = ["easily","walking","happiness","given","marked","inspecting"]

lemma_words = [word_lem.lemmatize(word) for word in words] # if you dont mention the pos tags, it cannot reduce the word to root form
lemma_words
```

```python
from nltk.stem import WordNetLemmatizer

word_lem2 = WordNetLemmatizer()

words = ["easily","walking","happiness","given","marked","inspecting"]

lemma_words2 = [word_lem.lemmatize(word,pos='v') for word in words]
lemma_words2
```

**Shallow Parsing**

```python
!python -m spacy download en_core_web_sm
```

```python
import spacy

# Load English small pipeline
nlp = spacy.load("en_core_web_sm")

data = '''
Data analysis is the process of inspecting, cleansing, transforming, and modeling data with the goal of discovering useful information, informing conclusions, and supporting decision-making.
Data analysis has multiple facets and approaches, encompassing diverse techniques under a variety of names, and is used in different business, science, and social science domains.
In today's business world, data analysis plays an important role in making decisions more scientific and helping businesses operate more effectively.
It is widely used in fields such as business analytics, healthcare, and artificial intelligence to extract meaningful insights from data.
'''

# Preprocess data : convert data into word tokens
doc = nlp(data)

# Extract the chunks from the sentence

for chunk in doc.noun_chunks:
  print("NP: ",chunk.text, "| Head: ",chunk.root.text, "|Label ",chunk.root.dep_)
```

**Deep Parsing**

```python
from spacy import displacy

# look at the deep parsing
for chunk in doc:
  print(f"{chunk.text:<15} {chunk.dep_} {chunk.head.text} {chunk.pos_}")
```

```python
# Visualize the above syntactic parsing
displacy.render(list(doc.sents)[0], style="dep", jupyter=True, options={"distance": 120})
```

### Transformer: Encoders and Decoders

```python
!python -m spacy download en_core_web_trf
```

```python
import spacy
nlp = spacy.load("en_core_web_trf")

text = '''Mark Zukerberg will meet Aditya Joshi
on Monday 25 November 2025, 10am for $3 Trillion deal
at Pune for second time.'''

doc = nlp(text)

for entity in doc.ents:
  print(entity.text,entity.label_)
```

```python
from spacy import displacy
displacy.render(doc,style='ent',jupyter=True)
```
