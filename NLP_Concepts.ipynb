## TOkenization

```python
data = '''
Data analysis is the process of inspecting, cleansing, transforming, and modeling data with the goal of discovering useful information, informing conclusions, and supporting decision-making.
[1] Data analysis has multiple facets and approaches, encompassing diverse techniques under a variety of names, and is used in different business, science, and social science domains.
[2] In today's business world, data analysis plays an important role in making decisions more scientific and helping businesses operate more effectively.
[3] It is widely used in fields such as business analytics, healthcare, and artificial intelligence to extract meaningful insights from data.
'''
print(data)
```

## Text cleaning

```python
# 1) convert all the data into lowercase
data = data.lower()

# 2) remove punctuation marks. remove numbers and also the special characters
import re
pattern = r'[^a-z. ]'
data = re.sub(pattern,'',data)
print(data)
```

## Divide the data into multiple tokens

Sentence Tokenization

```python
from nltk.tokenize import sent_tokenize
sent_tokens = sent_tokenize(data)
sent_tokens
```

```python
len(sent_tokens)
```

## Word tokenization

1) Unigrams
2) Bigrams : 2 words in a token
3) Trigrams : 3 words in a token
4) N grams : n words in a token

Unigrams : ['data', 'analysis', 'is', 'the', 'process', 'of',]

Bigrams : ['data analysis','analysis is','is the','the process','process of',..]

Trigrams : ['data analysis is','analysis is the','is the process','the process
of','process of ...]

N grams -5: ['data analysis is the process','analysis is the process of',...]


```python
bi_tokens = list(nltk.bigrams(word_tokens))
print(bi_tokens)
```

```python
tri_tokens = list(nltk.trigrams(word_tokens))
print(tri_tokens)
```

```python
ntokens = list(nltk.ngrams(word_tokens,5))
print(ntokens)
```

## If you are performing NLU, you can choose to remove stopwords. This will help in reducing the size of data

image.png

```python
from nltk.corpus import stopwords
stopwords.fileids()
```

```python
stop = stopwords.words('english')
print(stop)
print(len(stop))
```

## Removal of stopwords for NLU

```python
len(word_tokens)
```

```python
filtered_words = []
for word in word_tokens:
  if word not in stop:
    filtered_words.append(word)
print(filtered_words,len(filtered_words),sep='\n')
```

## Stemming and Lemmatization

Stemming

image.png

**Porter stemmer**

```python
from nltk.stem import PorterStemmer

words = ["easily","walking","happiness","given","marked","inspecting"]

ps = PorterStemmer()

ps_words = [ps.stem(word) for word in words]
print(ps_words)
```

**Snowball Stemmer** : supoorts stemming technique for other languages also

```python
from nltk.stem import SnowballStemmer

words = ["easily","walking","happiness","given","marked","inspecting"]

ss = SnowballStemmer("english")

ss_words = [ss.stem(word) for word in words]
print(ss_words)
```

**Lancaster Stemmer**

```python
from nltk.stem import LancasterStemmer

words = ["easily","walking","happiness","given","marked","inspecting"]

ls = LancasterStemmer()

ls_words = [ls.stem(word) for word in words]
print(ls_words)
```

## Lemmatization : brings the word to root form on the basis of part of speech tags

image.png

image.png



```python
from nltk.stem import WordNetLemmatizer

word_lem = WordNetLemmatizer()

words = ["easily","walking","happiness","given","marked","inspecting"]

lemma_words = [word_lem.lemmatize(word) for word in words] # if you dont mention the pos tags, it cannot reduce the word to root form
lemma_words
```

```python
from nltk.stem import WordNetLemmatizer

word_lem2 = WordNetLemmatizer()

words = ["easily","walking","happiness","given","marked","inspecting"]

lemma_words2 = [word_lem.lemmatize(word,pos='v') for word in words]
lemma_words2
```
